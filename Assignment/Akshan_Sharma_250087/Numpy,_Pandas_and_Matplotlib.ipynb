```python
import pandas as pd
```

```python
# Task 2.1: Create a DataFrame from the data of students above
data = {
    "Roll No.": [2301, 2310, 2302, 2304, 2306, 2305, 2307, 2312, 2315, 2316, 2320, 2308, 2309, 2314, 2319],
    "Name": ["Alice", "Smith", "Bob", "Charlie", "Eric", "Diana", "John", "Joey", "Bean", "Charlie",
             "Jack", "Jill", "George", "Tabitha", "Phoebe"],
    "Age": [24, None, 22, 23, None, 24, None, 30, 26, 23, 22, 21, 24, 31, 24],
    "Marks": [85, None, 78, 92, 95, 88, None, 64, 72, 92, 83, 69, 89, 100, 91]
}

df = pd.DataFrame(data)
```

***Hint:*** **NaN value** is represented by **None**

```python
# Task 2.2: Inspect the DataFrame(use the most basic functions to know the statistical quantities)
```

```python
# Task 2.3: Select the Marks column and Print it
print(df["Marks"])
```

```python
# Task 2.4: Access the second row
print(df.iloc[1])
```

```python
# (BONUS) Task 2.5: Access only the rows where Marks > 80
print(df[df["Marks"] > 80])
```

```python
# (BONUS, but really fun try to find it do it how :) ) Task 2.6: Make a function to assign grade
# (i) A if marks > 85
# (ii) B if marks > 75
# (iii) C if marks > 65
def assigngrade(marks):
  if marks > 85:
      return "A"
  elif marks > 75:
      return "B"
  elif marks > 65:
      return "C"
  else:
      return "D"
df["Grade"] = df["Marks"].apply(assigngrade)
```

```python
# Task 2.7: Rename Marks column to Score
df = df.rename(columns = {"Marks":"Score"})
print(df)
```

```python
# Task 2.8: Replace NaN in Age with mode and NaN in Marks with mean
df["Age"].fillna(df["Age"].mode()[0])
df["Score"].fillna(df["Score"].mean())
print(df)
```

```python
# Task 2.9: Remove duplicates
#There are no duplicates in the roll no but here you go
df = df.drop_duplicates(subset = ["Roll No."], keep = "first")
print(df)
```

```python
# Task 2.10: Group by the data by Grade and print the data
grouped = df.groupby("Grade")
for grade, data in grouped:
    print(f"\nGrade {grade}:\n", data)
```

```python
# Task 2.11: Sort Data in increasing order of Age
print(df.sort_values(by="Age", ascending=True))
```

```python
# Task 2.12: Export the dataframe to CSV
df.to_csv("export.csv")
```

```python
# Task 2.13: Import the given countries CSV
dataframe = pd.read_csv("export.csv")
print(dataframe)
```

# ***Part 3: Matplotlib***

```python
import matplotlib.pyplot as plt
```

```python
# Task 3.1: Plot a Pie Chart depicting the percentag of grades that the students have got.
grade_Data = list(df["Grade"])
a_p = grade_Data.count("A")/len(grade_Data)*100
b_p = grade_Data.count("B")/len(grade_Data)*100
c_p = grade_Data.count("C")/len(grade_Data)*100
d_p = grade_Data.count("D")/len(grade_Data)*100
plt.pie([a_p,b_p,c_p,d_p], labels = ["A","B","C","D"])
plt.show()
```

```python
# Task 3.2: Plot a Line Graph of Roll No. vs Marks
rolls = list(df["Roll No."])
scores = list(df["Score"])
acts = [(rolls[i],scores[i]) for i in range(len(rolls))]
acts.sort()
rolls = [k[0] for k in acts]
scores = [k[1] for k in acts]
plt.plot(rolls,scores)
plt.show()
```

```python
# Task 3.3: Plot a Histogram of for showing the Age distribution
plt.hist(list(df["Age"]))
plt.show()
```
