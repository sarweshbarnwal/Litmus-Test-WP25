```python
mat_1 = np.random.randint(0, 100, size=(3,3))
print("Matrix 1: \n", mat_1)
mat_2 = np.random.randint(0, 100, size=(3,3))
print("Matrix 2: \n", mat_2)
print("Addition: \n", mat_1 + mat_2)
print("Subtraction: \n", mat_1 - mat_2)
print("Multiplication: \n", mat_1 * mat_2)
total = 0
for i in range(3):
    for j in range(3):
        total += mat_1[i][j]
print("Sum with traversal:", total)
print("Sum without traversal:", np.sum(mat_1))
print("Mean: \n", np.mean(mat_1))
```

```python
import pandas as pd
```

```python
# Task 2.1: Create a DataFrame from the data of students above
```

```python
data = {
    "Roll No.": [2301, 2310, 2302, 2304, 2306, 2305, 2307, 2312, 2315, 2316, 2320, 2308, 2309, 2314,2319],
    "Name": ["Alice", "Smith", "Bob", "Charlie", "Eric", "Diana", "John", "Joey", "Bean", "Charlie", "Jack", "Jill", "George", "Tabitha", "Phoebe"],
    "Age": [24, np.nan, 22, 23, np.nan, 24, np.nan, 30, 26, 23, 22, 21, 24, 31, 24],
    "Marks": [85, np.nan, 78, 92, 95, 88, np.nan, 64, 72, 92, 83, 69, 89, 100, 91]
}
df = pd.DataFrame(data)
df
```

***Hint:*** **NaN value** is represented by **None**

```python
# Task 2.2: Inspect the DataFrame(use the most basic functions to know the statistical quantities)
```

```python
df.describe()
```

```python
print(df[df["Marks"] > 80])
```

```python
# (BONUS, but really fun try to find it do it how :) ) Task 2.6: Make a function to assign grade
# (i) A if marks > 85
# (ii) B if marks > 75
# (iii) C if marks > 65
```

```python
grades = []

for x in df["Marks"]:
    if x > 85:
        grades.append("A")
    elif x > 75:
        grades.append("B")
    elif x > 65:
        grades.append("C")
    else:
        grades.append("NaN")

df["Grade"] = grades
print(df)
```

```python
# Task 2.7: Rename Marks column to Score
```

```python
df.rename(columns = {"Marks": "Score"}, inplace = True)
print(df)
```

```python
# Task 2.8: Replace NaN in Age with mode and NaN in Marks with mean
```

```python
age_mode = df["Age"].mode()[0]
marks_mean = df["Score"].mean()
for i in range(len(df)):
    if pd.isna(df.iloc[i, df.columns.get_loc("Age")]):
        df.iloc[i, df.columns.get_loc("Age")] = age_mode
    if pd.isna(df.iloc[i, df.columns.get_loc("Score")]):
        df.iloc[i, df.columns.get_loc("Score")] = marks_mean
print(df)
```

```python
# Task 2.9: Remove duplicates
```

```python
df.drop_duplicates(subset="Name", inplace=True)
print(df)
```

```python
# Task 2.10: Group by the data by Grade and print the data
```

```python
for grade, data in df.groupby("Grade"):
    print("\nGrade:", grade)
    print(data)
```

```python
# Task 2.11: Sort Data in increasing order of Age
```

```python
df.sort_values(by="Age", inplace=True)
print(df)
```

```python
# Task 2.12: Export the dataframe to CSV
```

```python
df.to_csv("dataframe.csv")
```

# ***Part 3: Matplotlib***

```python
import matplotlib.pyplot as plt
```

```python
# Task 3.1: Plot a Pie Chart depicting the percentag of grades that the students have got.
```

```python
grade_counts = df["Grade"].value_counts()
grade_percent = (grade_counts / grade_counts.sum()) * 100
plt.pie(grade_percent, labels=grade_percent.index)
plt.title("Percentage of Grades Obtained by Students")
plt.show()
```

```python
# Task 3.2: Plot a Line Graph of Roll No. vs Marks
```

```python
plt.plot(df["Score"], df["Roll No."])
plt.xlabel("Marks")
plt.ylabel("Roll no.")
plt.title("Roll No. vs Marks")
plt.show()
```

```python
# Task 3.3: Plot a Histogram of for showing the Age distribution
```

```python
plt.hist(df["Age"])
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.title("Age Distribution of Students")
plt.show()
```
