Write a function count_vowels(text) that counts and returns the number of vowels (a, e, i, o, u) in a given sentence.

```python
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
       if char in vowels:
        count += 1
    return count
sentence = input("Enter a sentence: ")
print("Number of vowels: ", count_vowels(sentence))
```

Write a function longest_word(sentence) that returns the longest word in a given sentence

```python
def longest_word(sentence):
    words = sentence.split()
    longest = ""
    for word in words:
        if len(word) > len(longest):
            longest = word
    return longest
text = input("Enter a sentence: ")
print("The longest word is: ", longest_word(text))
```

Write a function reverse_string(text) that takes a string as input, reverses the string manually without using [::-1] and returns the reversed string.

```python
def reverse_string(text):
    reversed_text = ""
    for char in text:
        reversed_text = char + reversed_text
    return reversed_text
word = input("Enter a string: ")
print("Reversed string: ", reverse_string(word))
```

Write a function is_palindrome(word) that checks if a word is a palindrome.

```python
def is_palindrome(word):
    word = word.lower()
    reversed_word = ""
    for char in word:
        reversed_word = char + reversed_word
    return word == reversed_word
word = input("Enter a word: ")
if is_palindrome(word):
    print("It is a palindrome")
else:
    print("It is not a palindrome")
```

Sort this dictionary:

scores = {"A": 50, "B": 10, "C": 40, "D": 30}

Without using sorted() directly.

```python
scores = {"A": 50, "B": 10, "C": 40, "D": 30}
sorted_dic = {}
while len(sorted_dic) != len(scores):
 for k in scores:
  if k not in sorted_dic:
    min_key = k
    break
 for k in scores:
   if k not in sorted_dic and scores[k] < scores[min_key]:
    min_key = k
 sorted_dic[min_key] = scores[min_key]

print(sorted_dic)
```

Case Study: Smart Farmer ‚Äì Plant Leaf Disease Advisor

A group of farmers is using a mobile app called SmartLeaf.
The app allows them to click a photo of a plant leaf, and then it automatically analyses the image and converts it into a simple Python dictionary with the following keys:

1. "spots" ‚Üí color of spots on the leaf (e.g., "none", "brown", "black")

2. "patches" ‚Üí color of patches on the leaf (e.g., "none", "yellow", "white")

3. "holes" ‚Üí number of holes eaten by insects (integer)

4. "moisture" ‚Üí "low" or "normal"

Example of one leaf report generated from the image:

leaf_report = {
    "spots": "black",
    "patches": "yellow",
    "holes": 3,
    "moisture": "low"
}


The agricultural experts have given the following simple rules to detect plant problems:

1. If there are black spots and holes ‚â• 2 ‚Üí disease = "Leaf Blight"

2. Else if there are yellow patches and moisture is low ‚Üí disease = "Nutrient Deficiency"

3. Else if holes ‚â• 4 and there are no spots ‚Üí disease = "Insect Attack"

4. Otherwise ‚Üí disease = "Healthy"

Also, they want to assign severity level as:

1. "High" ‚Üí if disease is "Leaf Blight"

2. "Medium" ‚Üí if disease is "Nutrient Deficiency" or "Insect Attack"

3. "Low" ‚Üí if "Healthy"

For each disease, they also want to suggest a recommended action:

- "Leaf Blight" ‚Üí "Spray fungicide and remove infected leaves"

- "Nutrient Deficiency" ‚Üí "Add nitrogen-rich fertilizer and water regularly"

- "Insect Attack" ‚Üí "Use organic insecticide (e.g., neem oil)"

- "Healthy" ‚Üí "No action needed"

Q1) Write a Python function diagnose_leaf(leaf_report) that:

- takes a single leaf report (a dictionary) as input

- applies the above rules

- returns a new dictionary in the following format:

{
    "Disease": <disease_name_as_string>,
    "Severity": <severity_as_string>,
    "RecommendedAction": <action_as_string>
}


```python
def diagnose_leaf():
  leaf = {}
  result = {}
  leaf["spots"] = input( "color of spots on the leaf (e.g., none, brown, black): ")
  leaf["patches"] = input("color of patches on the leaf (e.g., none, yellow, white): ")
  leaf["holes"] = int(input("Number of holes eaten by insects: "))
  leaf["moisture"] = input("Moisture of leaves (low or normal): ")
  if(leaf["spots"]=="black" and leaf["holes"]>=2):
    result["Disease"] = "Leaf Blight"
  elif (leaf["patches"] == "yellow" and leaf["moisture"]=="low"):
    result["Disease"] = "Nutrient Deficiency"
  elif (leaf["spots"] == "none" and leaf["holes"]>=4):
    result["Disease"] = "Insect Attack"
  else:
    result["Disease"] = "Healthy"

  if (result["Disease"] == "Leaf Blight"):
    result["Severity"] = "High"
    result["RecommendedAction"] = "Spray fungicide and remove infected leaves"
  elif (result["Disease"] == "Nutrient Deficiency"):
    result["Severity"] = "Medium"
    result["RecommendedAction"] = "Add nitrogen-rich fertilizer and water regularly"
  elif (result["Disease"] == "Insect Attack"):
    result["Severity"] = "Medium"
    result["RecommendedAction"] = "Use organic insecticide (e.g., neem oil)"
  else:
    result["Severity"] = "Low"
    result["RecommendedAction"] = "No action needed"
  return result

result = diagnose_leaf()
print(result)
```

Q2) The app processes many leaf images from one field and stores them in a list of dictionaries, for example:

field_leaves = [
    {"spots": "black", "patches": "yellow", "holes": 3, "moisture": "low"},
    {"spots": "none", "patches": "yellow", "holes": 1, "moisture": "low"},
    {"spots": "none", "patches": "none", "holes": 5, "moisture": "normal"},
    {"spots": "brown", "patches": "none", "holes": 0, "moisture": "normal"}
]

Write Python code that:

- Uses the diagnose_leaf() function for each leaf

- Counts how many leaves fall under each disease category using a dictionary, for example:

{
    "Leaf Blight": 1,
    "Nutrient Deficiency": 1,
    "Insect Attack": 1,
    "Healthy": 1
}


Prints a clear summary report like:

Leaf Blight: 1 leaves

Nutrient Deficiency: 1 leaves

Insect Attack: 1 leaves

Healthy: 1 leaves

```python
def diagnose_leaf(leaf):
  result = {}
  if(leaf["spots"]=="black" and leaf["holes"]>=2):
    result["Disease"] = "Leaf Blight"
  elif (leaf["patches"] == "yellow" and leaf["moisture"]=="low"):
    result["Disease"] = "Nutrient Deficiency"
  elif (leaf["spots"] == "none" and leaf["holes"]>=4):
    result["Disease"] = "Insect Attack"
  else:
    result["Disease"] = "Healthy"

  if (result["Disease"] == "Leaf Blight"):
    result["Severity"] = "High"
    result["RecommendedAction"] = "Spray fungicide and remove infected leaves"
  elif (result["Disease"] == "Nutrient Deficiency"):
    result["Severity"] = "Medium"
    result["RecommendedAction"] = "Add nitrogen-rich fertilizer and water regularly"
  elif (result["Disease"] == "Insect Attack"):
    result["Severity"] = "Medium"
    result["RecommendedAction"] = "Use organic insecticide (e.g., neem oil)"
  else:
    result["Severity"] = "Low"
    result["RecommendedAction"] = "No action needed"
  return result
field_leaves = [
    {"spots": "black", "patches": "yellow", "holes": 3, "moisture": "low"},
    {"spots": "none", "patches": "yellow", "holes": 1, "moisture": "low"},
    {"spots": "none", "patches": "none", "holes": 5, "moisture": "normal"},
    {"spots": "brown", "patches": "none", "holes": 0, "moisture": "normal"}
]
disease_number = {}
for leaf in field_leaves:
  result = diagnose_leaf(leaf)
  disease = result["Disease"]
  if disease not in disease_number:
        disease_number[disease] = 0

  disease_number[disease] += 1
  for disease, count in disease_number.items():
    print(f"{disease}: {count} leaves")
```

Q3) Based on the summary, if the number of non-healthy leaves (i.e., total leaves minus healthy leaves) is more than half the total leaves, print:

"ALERT: Field at HIGH RISK. Call agricultural expert."

Otherwise, print:

"Field condition acceptable. Monitor regularly."

```python
total = sum(disease_number.values())
healthy_leaves = disease_number.get("Healthy", 0)
not_healthy = total - healthy_leaves
if not_healthy > total / 2:
  print("ALERT: Field at HIGH RISK. Call agricultural expert.")
else:
  print("Field condition acceptable. Monitor regularly.")
```
